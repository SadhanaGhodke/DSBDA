```python
from sklearn.preprocessing import MinMaxScaler

features = df.iloc[:, :-1]  # All columns except the last one (numerical features)
labels = df.iloc[:, -1]     # Last column (categorical labels)

scaler = MinMaxScaler()
features_scaled = pd.DataFrame(scaler.fit_transform(features), columns=features.columns)

df_scaled = pd.concat([features_scaled, labels], axis=1)

print(df_scaled.head())
```

```python
from sklearn import preprocessing
label2 = preprocessing.LabelEncoder()
df['en_label'] = label2.fit_transform(df['label'])
df['en_label'].unique()
```

```python
df['dec_label'] = label2.inverse_transform(df['en_label'])
df['dec_label'].unique()
```
