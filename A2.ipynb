```python
col = ['Math_Score', 'Reading_Score', 'Writing_Score', 'Placement_Score']
df.boxplot (col)
```

```python
print (np.where (df ['Math_Score'] < 40))
print (np.where (df ['Reading_Score'] < 60))
print (np.where (df ['Writing_Score'] < 40))
print (np.where (df ['Placement_Score'] < 50))
```

```python
from scipy import stats
z = np.abs (stats.zscore (df ['Math_Score']))
print (z)
```

```python
threshold = 2.00
outliers = np.where (z > threshold)
outliers
```
