```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/iris_synthetic_data.csv')
df.head()
```

```python
df.describe()
```

```python
group = df.groupby('label')
# print(group)
group.describe()
```

```python
print("Sepal Length Mean :", df["sepal length"].mean())
print("Sepal Length Median :", df["sepal length"].median())
print("Sepal Length Mode :", df["sepal length"].mode())
print("Sepal Length Standard Deviation :", df["sepal length"].std())
print("Sepal Length Minimum :", df["sepal length"].min())
print("Sepal Length Maximum :", df["sepal length"].max())
```

```python
range = group.max() - group.min()
print ("Range:\n", range)
std = group.std()
print ("\nStandard Deviation:\n", std)
mean = group.mean()
print ("\nMean:\n",mean)
median = group.median()
print ("\nMedian:\n",median)
variance =group.var()
print ("\nVariance:\n",variance)
```

```python
Q1 = 0.25
Q2 = 0.50
Q3 = 0.75
quartiles = group.quantile([Q1,Q2, Q3])
print("Quartiles (Q1, Q2, Q3) of each feature by label:")
print(quartiles)
IQR = group.quantile(Q3) - group.quantile(Q1)
print("\nInterquartile Range (IQR) of each feature by label:")
print(IQR)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
print("\n\n")
sns.boxplot(x='label', y='sepal length', data=df)
plt.title('Boxplot of Sepal Length by Species')
plt.show()
print("\n\n")
sns.boxplot(x='label', y='sepal width', data=df)
plt.title('Boxplot of Sepal Width by Species')
plt.show()
print("\n\n")
sns.boxplot(x='label', y='petal length', data=df)
plt.title('Boxplot of Petal Length by Species')
plt.show()
print("\n\n")
sns.boxplot(x='label', y='petal width', data=df)
plt.title('Boxplot of Petal Width by Species')
plt.show()
print("\n\n")
```

```python
plt.figure(figsize=(12, 8))

sns.histplot(data=df, x='sepal length', hue='label', kde=True, bins=15, multiple="stack")

plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length by Species')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.histplot(data=df, x='sepal width', hue='label', kde=True, bins=15, multiple="stack")

plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width by Species')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.histplot(data=df, x='petal width', hue='label', kde=True, bins=15, multiple="stack")

plt.xlabel('Petal Width')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Width by Species')

plt.show()
```
